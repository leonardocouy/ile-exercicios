### Exercício 5.1 

Escreva um programa que leia números repetidamente até que o usuário informe “done”. Uma vez que “done” é informado, imprima a soma, quantidade e média dos números. Se o usuário digita alguma coisa diferente de um número, detecte o erro usando try e except e imprima uma mensagem de erro e pule para o próximo número.

````python
# Numbers list
numbers = []

while True:
    
    try:
        result = input("Digite um número: ")
        if isinstance(int(result), int):
            numbers.append(int(result))
    except ValueError:
        if result != "done":
            print("Invalid Input")
        else:
            numbers_sum = sum(numbers)
            numbers_quant = len(numbers)
            numbers_mean = numbers_sum / numbers_quant
            print("S: {n_sum} Q: {quant} M: {mean}".format(n_sum=numbers_sum, quant=numbers_quant, mean=numbers_mean))
            break
````

### Exercício 7.2

Escreva um programa para perguntar por um nome de arquivo, e então, ler o arquivo e procurar por linhas da forma:

Quando você encontra uma linha que começa com “X-DSPAM-Confidence:” separe a linha para extrair dela o número com ponto flutuante. Conte essas linhas e determine o valor total médio dos valores de confiança de spam a partir destas linhas. Quando você alcançar o fim do arquivo, imprima a confiança média de spam.

````python
filename = input('Digite o nome do arquivo: ')

spam_confidences = []

with open(filename, 'r') as f:
    text = f.readlines()
    for line in text:
        if "X-DSPAM-Confidence: " in line:
            # extract value and transform to float
            spam_confidence = float(line.strip('X-DSPAM-Confidence: '))
            
            # append to spam confidences list.
            spam_confidences.append(spam_confidence)

mean = sum(spam_confidences) / len(spam_confidences)
print('A média da confiança de spam é: {mean}'.format(mean=mean))
````

### Exercício 10.3
Escreva um programa que leia um arquivo e imprima as letras em ordem decrescente de frequência. Seu programa deve converter todas as entrada para minúsculo e somente contar as letras de a-z. Seu programa não deve contar espaços, dígitos, pontuação ou qualquer outros que não sejam as letras de a-z. Encontre exemplos de textos de diferentes linguagens e veja como a frequência das letras variam entre as linguagens. Compare seu resultado com as tabelas em wikipedia.org/wiki/Letter_frequencies.

````python
import re
import string

filename = input('Digite o nome do arquivo: ')
char_ocurrences = {}

with open(filename, 'r') as f:
    text = f.read().lower()
    
    # Regular Expression to strip all special characters
    text = re.sub('[^a-z]+', '', text)
    
    # Get all characters from alphabet
    alphabet = list(string.ascii_lowercase)
    print(alphabet)
    
    # Iterate and count
    for char in alphabet:
        char_ocurrences[char] = text.count(char)
    
    # Sort by Descending
    char_ocurrences = sorted(char_ocurrences.items(), key=lambda char: char[0], reverse=True)
    print(char_ocurrences)
    
    # Pretty print
    for char, freq in char_ocurrences:
        print('{char} - {freq}'.format(char=char, freq=freq))
````
